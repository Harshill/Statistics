```python
import numpy as np
import pymc3 as pm
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as graph

graph.style.use('fivethirtyeight')
```

```python
data = stats.norm.rvs(loc=100, scale=15, size=56)

sns.distplot(data)
graph.show()
```

```python
with pm.Model() as mean:
    mu = pm.Normal('mu', 100, 100)
    sd = pm.HalfNormal('sd', 100)
    
    obs = pm.Normal('obs', mu, sd, observed=data)
    
    trace = pm.sample(5000)
    
    pm.traceplot(trace)
    graph.show()
```

# Linear Regression

```python
m, b = -5, -1

x = np.random.rand(55) * 5
y = m * x + b + 4 * np.random.randn(len(x))

graph.plot(x, y, 'o')
graph.show()
```

```python
with pm.Model() as reg:
    # Priors
    b_est = pm.Normal('b_est', 0, 100)
    m_est = pm.Normal('m_est', 0, 100)
    sd = pm.HalfNormal('sd', 100)
    
    # Likelihood
    obs = pm.Normal('obs', m_est * x + b_est, sd=sd, observed=y)
    
    # Sample
    trace = pm.sample(1000)
    
    # Results
    pm.traceplot(trace, varnames=['b_est', 'm_est'])
    graph.show()
    
    # Posterior
    pm.plot_posterior(trace, point_estimate='mean', varnames=['b_est', 'm_est'], ref_val=[b, m])
    graph.show()
```
